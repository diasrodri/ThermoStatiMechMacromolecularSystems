
## 1.2.1 A tríade de sequência de aminoácidos, estrutura e função

> As proteínas são macromoléculas altamente especializadas, voltadas para o controle de sistema biológico tais como; processos de transporte, estabilização da estrutura celular, e catalisação enzimática de reações químicas; outros agem como motores moleculares em uma maquinaria complexa de processos de sintetização molecular.

> Quimicamente, proteinas são construídas por seqüências de resíduos de aminoácidos ligados por ligações peptídicas. 

>A cadeia polipeptídica consiste de uma coluna principal linear com cadeias laterais de aminoácidos específicos ligados a ele.

As cadeias laterais destes aminoácidos governam a especificidade principal de cada aminoácido nos processos de dobragem de proteínas e diferem nas propriedades químicas e físicas sob a influência do solvente circundante.

A solubilidade num ambiente aquoso é dependente da ocorrência de grupos polares na cadeia lateral, tais como, por exemplo, os grupos hidroxílicos em serina e treonina. 

Cadeias laterais hidrofóbicas são insolúveis e pouco reativas em um ambiente polar. 

Cadeias laterais grandes e fortemente hidrófobas típicas, tais como, por exemplo, fenilalanina ou triptofano, possuem anéis aromáticos.

Outros, como alanina ou leucina, com grupos methine (-CH), methylene (-CH2), ou methyl (-CH3) na cadeia lateral, são alifáticos, i.e., estas cadeias laterais somente contém hidrogênio e carbono.

Principalmente apenas arginina e lisina (carregada positivamente) e ácido aspártico e glutâmico (carregada negativamente) contribuem explicitamente para a carga total de uma proteína em um ambiente neutro. Além disso, a histidina é tipicamente carregada positivamente em um ambiente levemente ácido, mas neutra em solução neutra.



A frequência e disposição sequencial dos resíduos de aminoácidos hidrofóbicos, polares e carregados na sequência de aminoácidos (também designada por *estrutura primária*) de uma proteína é principalmente responsável pela formação de uma conformação nativa estável e única.

Conformações proteicas ou segmentos dela são tipicamente classificadas em diferentes escalas de comprimento.

Partes de conformações de proteínas que formam subestruturas simétricas locais são chamadas de *estruturas secundária*s.

Estes incluem hélices, folhas planas (ou cordões) e voltas. Subestruturas deste tipo são comuns a todos os objetos semelhantes a linhas e podem ser geralmente consideradas como a geometria subjacente de polímeros lineares.

A formação de estruturas secundárias não está necessariamente conectada com a formação de ligações de hidrogênio, mas essas estruturas são essencialmente estabilizadas por ligações de hidrogênio.

Toda a conformação de uma única proteína, incluindo estruturas secundárias alinhadas, define sua *estrutura terciária*.

A estrutura terciária consiste tipicamente num núcleo muito compacto de resíduos hidrofóbicos que é blindado do ambiente aquoso por um invólucro de aminoácidos polares estando em contacto directo com o solvente.

### Esse conjunto de partes polares e hidrofóbicas separadas é característico de proteínas - isto reduz a entropia e, portanto, garante a estabilidade. Eventualmente, em grandes proteínas ou compostos proteicos, diferentes domínios hidrofóbicos podem formar-se independentemente.

A forma global da macromolécula que pode incluir compósitos de domínios terciários individuais é geralmente classificada como estrutura quaternária [1-3].

### A compreensão dos aspectos gerais do enrolamento de proteínas em conformações nativas é particularmente essencial, pois a forma estável da estrutura geométrica tridimensional de uma proteína freqüentemente determina a função biológica de uma proteína - ou seu mau funcionamento, se a proteína tiver desdobrado, redobrados ou desnaturados.

O espectro de funções proteicas em um organismo biológico é múltiplo.

Proteínas estão envolvidas em quase todos os processos celulares. Canais iônicos e nanoporos formados por proteínas de membrana controlam o íon e o fluxo de água para dentro e para fora da célula [4].

A eficiência é extrema. Até 3 000 000 000 moléculas de água podem atravessar este poro por segundo [5,6].


### Outras proteínas celulares como a actina, por exemplo, são responsáveis pela estabilidade mecânica do esqueleto celular.

### Actina polimeriza filamentos e estes filamentos podem formar redes estáveis.

Além da estabilidade das células, estas redes também permitem processos de transporte ao longo destas “pistas”, por exemplo, transporte de vesículas mediadas por proteínas de miosina.

A interação entre a actina e a miosina também é importante para a capacidade do tecido muscular de se contrair.

Reações bioquímicas são catalisadas por enzimas.

### A estabilidade geral da estrutural proteica, mas também a capacidade de se desdobrar e redobrar localmente, permitem processos de ligação *receptor-ligante*, o que é necessário para a actividade enzimática.

Grandes proteínas ou compostos proteicos formam máquinas complexas em nanoescala que atuam como “motores moleculares”, como, por exemplo, nas polimerases de DNA/RNA e sintase de ATP.
