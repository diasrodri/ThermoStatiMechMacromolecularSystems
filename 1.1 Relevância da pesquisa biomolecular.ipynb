# 1.1 Relevância da pesquisa biomolecular

Mais de 30 milhões de pessoas, infectadas pelo vírus da imunodeficiência humana (HIV), sofrem da síndrome da imunodeficiência adquirida (AIDS);

12 bilhões de seres humanos carregam o vírus da hepatite B (HBV) dentro de si e em mais de 350 milhões de casos a doença hepática causada pelo HBV é crônica e, portanto, atualmente incurável.

Os vírus consistem tipicamente num ácido nucleico compactamente dobrado (RNA ou DNA de cadeia simples ou dupla) encapsulado por um casco de proteína.

Uma vez que os processos celulares regulares são perturbados pela infecção do vírus, a consequência pode ser sérios danos ao organismo biológico ou mesmo a destruição da rede funcional do mesmo.



>Outra classe de doenças é devido a mudanças estruturais de proteínas mediadas por outras moléculas, chamados **prions**[link_text](https://pt.wikipedia.org/wiki/Pr%C3%ADon).

>Como existe uma forte conexão causal entre a estrutura tridimensional de uma proteína e sua função biológica, reenrolamento pode causar a perda de sua funcionalidade. Uma conseqüência possível é a morte das células.

>Exemplos de doenças priônicas no cérebro são encefalopatia espongiforme bovina (BSE)3 ou vulgamente conhecida por doença da vaca louca,  e sua forma humana doença de Creutzfeld-Jakob (CJD) 4.



Uma fonte adicional para danificar as redes celulares é o desdobramento da proteína seguido pela agregação amilóide. [link_text](https://run.unl.pt/bitstream/10362/16248/1/Falcao_2015.pdf)

No caso da doença de Alzheimer (AD), que é uma doença neurodegenerativa e o tipo mais comum de demência, os peptídeos beta-amilóide (Aβ) em concentração suficientemente alta experimentam mudanças estruturais e tendem a formar agregados. Seguindo a hipótese da amilóide, acredita-se que estes agregados (que também podem tomar formas fibrilares) são neurotóxicos, isto é, são capazes de se fundirem nas membranas celulares dos neurônios e abrir canais iônicos de cálcio. Sabe-se que os íons extracelulares de Ca2+ que se intruzem em um neurônio podem promover sua degeneração. Cerca de 24 milhões, principalmente idosos, são afetados atualmente.

>Esta coleção de doenças manifesta a extraordinária importância da pesquisa molecular biologicamente motivada.  

>Esta pode levar a compreensão da síntese de biomoléculas como as proteínas baseadas no código genético (expressão gênica), as características e especificidades dos processos de dobramento e agregação, bem como o desdobramento dos aspectos dinâmicos e cinéticos dos processos biológicos. 

>Em quase todos esses processos estão envolvidas proteínas funcionais. Evidentemente, tratar pacientes que sofrem dessas doenças é um problema médico, mas revelar a natureza dos mecanismos por trás do funcionamento dos organismos vivos é uma tarefa interdisciplinar para todo o conjunto de ciências naturais.

>Esse também é o caso quando se trata de aplicações nanotecnológicas. A ciência e a tecnologia se encontram e se fundem parcialmente em menore escalas de comprimento. 

>Em particular, aprender com sistemas biológicos - a partir de redes complexas de funcionalidade biológica ou de propriedades de moléculas únicas - tornou-se cada vez mais importante e relevante para aplicações especiais em escalas micrômetricas ou nanométricas. Isso também inclui a interação de tais materiais macios com matéria sólida, onde a pesquisa sistemática está apenas no começo.

>Exemplos de tais sistemas híbridos com enorme potencial para futuras aplicações são, entre muitos outros, biosensores em forma de nanoarranjos específicos para a identificação de proteínas em solução e também circuitos nanoeletrônicos baseados em polímeros. 

> Do lado experimental, o progresso no desenvolvimento de equipamentos de alta resolução e novas técnicas experimentais não só permite detectar o que está acontecendo em escalas atômicas, mas também permite a manipulação local de moléculas, o que é essencial para se projetar aplicações específicas.

>Por outro lado, as capacidades computacionais atingiram um nível que possibilita o estudo de sistemas macromoleculares de forma mais sistemática em simulações de modelos adequados por meio de métodos numéricos sofisticados. Simulações são particularmente relevantes para investigações de processos que ainda estão inacessíveis
para pesquisa experimental.




>Neste ponto, o desafio para a física teórica, em particular, é duplo em primeiro lugar, a modelagem e análise de estruturas moleculares específicas em escalas atômicas e em segundo lugar, a generalização de transições da estrutura conformacional (pseudofase) que acompanha processos de formação dentro de um quadro mesoscópico. Ambas as abordagens facilitam a compreensão sistemática de processos moleculares que são tipicamente difíceis de alcançar em experimentos. 

>Dobramento de proteínas, agregação de peptídeos, colapso de polímeros, cristalização e adsorção de polímeros e proteínas para nanopartículas e substratos sólidos têm uma característica essencial em comum:

####Em todos esses processos, a formação de estruturas é guiada por um  comportamento coletivo e cooperativo das subunidades moleculares que se alinham para construir macromoléculas semelhantes a cadeias.

>Neste processo, polímeros e proteínas experimentam transições conformacionais relacionadas a transições de fase termodinâmicas. Para cadeias de comprimento finito, uma diferença importante de cruzamentos entre fases (pseudo) conformacionais é, no entanto, que essas transições são processos tipicamente bastante suaves, isto é, a actividade termodinâmica não é necessariamente sinalizada por fortes flutuações entrópicas ou energéticas.

O interesse nas propriedades de polímeros de comprimento finito e proteínas tem crescido rapidamente nos últimos anos, não apenas por causa dos avanços tecnológicos em escala nanométrica, mas em particular devido ao fato de que a termodinâmica de sistemas de pequena escala é a chave para a compreensão dos princípios biomecânicos da troca de sinais e dos processos de transporte, sendo relevante para a vida, como exemplo, ligação receptor-ligante entre proteínas ou fluxo molecular através de nanoporos.

Como as proteínas desempenham um papel fundamental em todos os processos biológicos, discutiremos suas propriedades mais especificamente nas proximas seções.
