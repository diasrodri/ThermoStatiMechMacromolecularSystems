# 1.2.2 Síntese ribossômica de proteínas

A informação sobre a composição de aminoácidos das proteínas é codificada no DNA.

No processo de polimerase II, o código genético é transcrito para a sequência de RNA (mRNA) mensageiro de cadeia simples.

A tradução do código de base do mRNA em sequências de aminoácidos, isto é, a expressão do gene, é mediada pelo ribossoma.

O próprio ribossomo é um composto macromolecular de grandes proteínas de domínio múltiplo. Um instantâneo esquemático da síntese proteica ribossômica é mostrado na Fig. 1.4.

Sempre três bases sucessivas ao longo da cadeia de mRNA codificam um único aminoácido. O tamanho de um codon é intuitivamente claro:

Uma vez que existem quatro bases diferentes construindo o código de RNA e a base complementar necessária para formar um par de bases é única, um codão de base simples poderia codificar apenas 4 aminoácidos e duas bases $4^2 = 16$ resíduos de aminoácidos.
 
Como 20 aminoácidos foram identificados em bioproteínas típicas, são necessárias três bases para formar um códon.

Como $4^3 = 64$ possibilidades não só permitem múltiplos códigos redundantes para aminoácidos (que é um primeiro tipo de correção de erro genético), mas também permitem a definição de códons de partida e parada que são necessários para separar os códigos para as diferentes proteínas dentro da sequência de RNA linear [7,8].

.








Após o ribossoma ter lido um codão da cadeia de mRNA, uma molécula de RNA de transferência (tRNA) liga-se ao codão de mRNA.

Uma molécula de tRNA contém principalmente uma secção de três bases, o anticódon, que é complementar de um codão específico no mRNA, e o resíduo de aminoácido associado.

Assim, as moléculas de tRNA servem como tradutores entre os codões de base e os aminoácidos.

O ribossoma separa o aminoácido do tRNA e o liga à parte já sintetizada da seqüência proteica.

Esse processo continua até que um códon de parada seja atingido.

Eventualmente, a proteína é liberada no solvente aquoso dentro da célula. 

Acredita-se que neste momento a proteína ainda não está estruturada e a formação da estrutura funcional é um processo de dobramento espontâneo.

Proteínas maiores que exibem uma tendência aumentada para dobrar no ambiente complexo e cheio são freqüentemente encapsuladas em acompanhantes que auxiliam no processo de dobramento.
