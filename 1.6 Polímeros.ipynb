# 1.6 Polímeros




### Bioproteins form a large class of molecules, but are only a subset of the much larger class of polymers yet.

### Other biologically relevant groups of molecules include the single- and double-stranded deoxyribonucleic acids (DNA) and ribonucleic acids (RNA), the carriers and messengers of genetic information, respectively.

### Other classes accommodate synthetic peptides and polymers which have substantial relevance in nanotechnology.
