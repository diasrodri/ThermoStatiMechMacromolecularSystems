# 1.6.4 Polímeros elásticos


### For some classes of polymers, under certain conditions, it is also necessary to model longitudinal bond vibrations.

### The elastic bond potential that describes the confinement of bonded monomers must be repulsive in both limits of too short and too large distances between the monomers.

### Fluctuations are only possible in a finite region about the equilibrium bond length $r_0$.

### Often, the bond is simply considered as a harmonic spring and, therefore, a harmonic approximation is used to model bond fluctuations (Rouse model),

\begin{equation}
V_{bond}^{harm}(r_{ii+1})=\frac{1}{2}K(r_{ii+1}-r_{0})^2,~~~~~ (1.37)
\end{equation}

### where r i i+1 is the distance between bonded monomers $i$ and $i + 1$ and $K$ is the spring constant. 

### For this reason, coarse-grained representations of elastic polymers are typically referred to as bead-spring models in contrast to bead-stick models for polymers, for which bond vibrations can be neglected (an example is the peptide bond in proteins).

### To explicitly take into account that the bond is confined and thus the fluctuation width is finite, the potential of bonded monomers can be introduced as


\begin{eqnarray}
V_{bond}^{sym}(r_{ii+1})=\left\{
                \begin{array}{ll}
                  V_{FENE}(r_{ii+1}) & if & |r_{ii+1}-r_0|\leq R, \\
                  \infty             & if & |r_{ii+1}-r_0|\gt R, 
                \end{array}
              \right.~~~~~ (1.38)
\end{eqnarray}


### The bond potential within the confinement region $|r_{ii+1}-r_0|\ll R$ (the total symmetric fluctuation width is $2R$ and centered about $r_0$) is typically modeled by the finitely extensible nonlinear elastic (FENE) potential [50], which we introduce here in the form [51]

\begin{equation}
V_{FENE}(r_{ii+1})=-\frac{1}{2}KR^2\ln(1-[(r_{ii+1}-r_{0})/R]^2,~~~~~ (1.39)
\end{equation}

### For small bond vibrations $(|r_{ii+1}-r_{0}|\ll R)$, the leading term of the Taylor expansion yields the harmonic potential (1.37).

### If one is interested in studies of the limit $K → 0$, all interactions between the monomers of the dissolved chain must be non-bonded interactions, which are typically modeled by the Lennard-Jones potential (1.34).

### Thus, a bonded Lennard-Jones potential is sometimes added to the FENE potential [52,53]:

\begin{equation}
V_{bond}(r_{ii+1})=V_{bond}^{sym}(r_{ii+1})+4\varepsilon\left[\left(\frac{\sigma}{r_{ii+1}}\right)^{12}-\left(\frac{\sigma}{r_{ii+1}}\right)^{6}\right],~~~~~ (1.40)
\end{equation}

### This bond potential is obviously asymmetric. Another frequently used variant is an asymmetric form of the bond potential that consists of the combination of the repulsive part of the Lennard-Jones potential and the attractive part of the FENE potential [50,54]:

\begin{eqnarray}
V_{bond}^{comb}(r_{ii+1})&=&\left\{
                \begin{array}{ll}
                  4\varepsilon\left[\left(\frac{\sigma}{r_{ii+1}}\right)^{12}-\left(\frac{\sigma}{r_{ii+1}}\right)^{6}+\frac{1}{4}\right] & if & r_{ii+1} \leq 2^{1/6}\sigma, \\
                  0             & if & r_{ii+1} \geq 2^{1/6}\sigma, 
                \end{array}
              \right.\\
              &+&\left\{
                \begin{array}{ll}
                  -\frac{1}{2}KR^2\ln(1-[(r_{ii+1}-r_{0})/R]^2 & if & r_{ii+1} \leq R_0, \\
                  \infty             & if & r_{ii+1} \geq R_0, 
                \end{array}
              \right.~~~~~ (1.41)
\end{eqnarray}

### All these variants of the bond potential are generic, which means that elasticity is captured as a property of chemical, covalent bonds, but this is done in a rather unspecific way.

### For most investigations of structural phases of polymers on coarse-grained level, details will not be relevant, although it should also be pointed out that surface properties of crystalline or amorphous structures of finite polymers are influenced by differences in inherent length scales of competing bonded and non-bonded interactions.

### Such length scales are given, for example, by respective locations of potential minima.

### Lattice simulations of polymers with variable bond length are done best by using the bond-fluctuation model [55,56].

### In three dimensions on a simple-cubic lattice, a monomer is placed in the center of a unit cell of this underlying lattice.

### Because of self-avoidance no other monomer can occupy the same site, i.e., the chain is not allowed to cross itself.

### The bond-fluctuation model is designed in a way that such crossings are inherently excluded.

### It can be shown that, if we introduce the basis set of 6 vectors

\begin{equation}
\vec{{\bf b}}^{(1)}=\left( \begin{array}{l} 2 \\ 0 \\ 0 \end{array} \right), \vec{{\bf b}}^{(2)}=\left( \begin{array}{l} 2 \\ 1 \\ 0 \end{array} \right), \vec{{\bf b}}^{(3)}=\left( \begin{array}{l} 2 \\ 1 \\ 1 \end{array} \right), \vec{{\bf b}}^{(4)}=\left( \begin{array}{l} 2 \\ 2 \\ 1 \end{array} \right), \vec{{\bf b}}^{(5)}=\left( \begin{array}{l} 3 \\ 0 \\ 0 \end{array} \right), \vec{{\bf b}}^{(6)}=\left( \begin{array}{l} 3 \\ 1 \\ 0 \end{array} \right) ~~~~~ (1.42)
\end{equation}

### each allowed bond vector between bonded monomers $i$ and $i + 1$ can only be one that belongs to this set:

\begin{equation}
\{\vec{r}_{ii+1}\}=\bigcup_{k=1}^{6}\mathscr{P} \vec{{\bf b}}^{(k)} ~~~~~ (1.43)
\end{equation}

### In this representation, $\mathscr{P}$ is an operator that allows all exchanges of spatial components and permutations of signs for each component of the vector it is applied to. 

### Thus, $\mathscr{P} \vec{{\bf b}}^{(k)}=\{(\pm b^{(k)}_l,\pm b^{(k)}_m,\pm b^{(k)}_n)\}$ , where $(l, m, n) = P(1, 2, 3)$ is any permutation of $\{1, 2, 3\}$.
### The set (1.43) contains 108 different bond vectors $(k = 1, 5: 6, k = 2, 3, 4, 6: 
24)$.
