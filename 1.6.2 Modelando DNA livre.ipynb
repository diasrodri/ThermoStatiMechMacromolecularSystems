# 1.6.2 Modelando DNA livre


### The simplest model for free (or linker) DNA is the wormlike-chain or Kratky-Porod model
[48].

### It is based on the assumption that changing the contour of a linear chain by bending costs energy.

### If we describe the contour of length $l$ by introducing the contour parameter $s∈[0,l]$, an infinitesimal segment of the contour (arc length) can be expressed in local coordinates by

\begin{equation}
ds^2=dx^2(s)+dy^2(s)+dz^2(s)
\end{equation}

### or short $ds=\sqrt{d\vec{x}^2(s)}$. The contour integral cam then be written as 

\begin{equation}
l=\int_0^l ds =\int_0^l ds \sqrt{\frac{d\vec{x}^2(s)}{ds^2}}=\int_0^l ds |\vec{u}(s)|
\end{equation}(1.16)

### Here we have introduced the tangent vector


\begin{equation}
\vec{u}(s)=\frac{d\vec{x}(s)}{ds}
\end{equation}

### which, according to Eq. (1.16), has unit length, $|\vec{u}| = 1$. The contour is bent at $s$, if

\begin{equation}
\frac{d\vec{u}(s)}{ds}=\lim_{\Delta s \rightarrow 0} \frac{\vec{u}(s+\Delta s)-\vec{u}(s)}{\Delta s}\ne0
\end{equation}

### or

\begin{equation}
\vec{u}(s+\Delta s)\cdot\vec{u}(s)=\cos\theta\ne 1
\end{equation}(1.19)


### i.e., the bending angle is non-zero: $\theta\ne0$.

### The left-hand side of Eq. (1.19) reminds us of the spin-spin interaction, as described by the Heisenberg model of a topologically one-dimensional, discrete chain of $n$ spin vectors $\vec{S}_i$, embedded into three-dimensional space, with $|\vec{S}_i|=1$:$E=-J\sum_{i=1}^{n-1}\vec{S}_i\cdot\vec{S}_{i+1}$, where $J$ is the coupling strength between neighboring spins.

### Spin alignment is obviously energetically favorable; bending along the spin chain costs energy. 

### We can transfer this idea to the polymer model, discretize the contour by splicing $n$ finite segments $\vec{b}_i$ of length $|\vec{b}_i|=b$ under the constraint $nb=l$, and re-express the left hand side of Eq. (1.19) by

\begin{equation}
-\vec{b}_i\cdot\vec{b}_{i+1}=\frac{1}{2}[(\vec{b}_{i+1}-\vec{b}_i)^2-2b^2]=-b^2\cos\theta_i
\end{equation}(1.20)

### If we fix the energy scale in a way that the entirely straight chain has zero bending energy, the discrete semiflexible, wormlike-chain polymer model can be written as

\begin{equation}
E_{bend}=\kappa\sum_{i=1}^{n-1}(1-cos\theta_i)=\frac{1}{2}\kappa b \sum_{i=1}^{n-1}b\left(\frac{\vec{b}_{i+1}-\vec{b}_i}{b}\right)^2
\end{equation}(1.21)

### where $\kappa$ is a material parameter associated with the bending stiffness of the discrete polymer.

### For the continuous formulation, we substitute $b \rightarrow \Delta s$, $\vec{b}_i \rightarrow  \vec{u}(s)$, $\vec{b}_{i+1} \rightarrow  \vec{u}(s+\Delta s)$,su ch that in the limit $\Delta s \rightarrow 0$ the sum in Eq. (1.21) can be replaced by an integral. 

### The continuous form of the wormlike-chain model then reads:

\begin{equation}
E_{bend}=\frac{1}{2}\bar{\kappa} \int_0^l ds\left[\frac{d\vec{u}(s)}{ds}\right]^2
\end{equation}(1.22)

### In this representation, the bending stiffness is given by $\kappa=\lim_{b→0,n→∞} \kappa$ (but still $nb=l=const$).

### Given the simplicity, it is rather surprising that various dynamic and structural properties of DNA and RNA can successfully be described by this model. 

### One essential feature is that the exponential decay of the bond-bond correlation function defines a characteristic length scale, the persistence length $\xi$, as we have already mentioned it above,

\begin{equation}
\langle \vec{u}(s) \vec{u}(s^{'}) \rangle \sim e^{-|s-s^{'}|/\xi}
\end{equation}(1.23)

### In three-dimensional embedding space, it is $\xi=\kappa/k_B T$ , where $k_B$ is the Boltzmann constant and $T$ is the temperature.

### Thus, since $\kappa$ is a material parameter and if $T$ is fixed (e.g., room temperature), this result implies that $\xi = const$, independent of $l$.

### Therefore, free DNA, RNA chains, and also some protein complexes such as myosin fibers and actin filaments, contained in the cytoskeleton of cells, can typically be well characterized and identified by their persistence length.
